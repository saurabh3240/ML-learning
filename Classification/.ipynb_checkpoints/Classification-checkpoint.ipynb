## Clasification of CLASSIFICATION on the bases of Data


### Binary Classifier
- only two classes (most often used)
- e.g. Titanic Dataset ( survived vs died)
   

### Multi-Class Classifier
- A classification task with more than two classes
- Assume that each sample is assigned to one and only one label
- e.g. given a flowers features identify its species (IRIS dataset)
- here classes are mutually exclusive

### Multi-Label Classifier
- assigns to each sample a set of target labels
- classes are not mutually exclusive
- e.g. Given an image identify the objects in that image

![errors](1.png)

### confusion matrix
|Test\Actual| True | False|
|-------|----|----|
|True|True Positve (TP)|false positive (FP)|
|False|False Negative (FN)|true negative (TN)|



**Type I error **
- predictive positive but acutally false (FP). 
- incorrect rejection of true null hypothesis ( do no be confused)
- p value test 
	- value is very small $\mu \pm 2*\sigma$ or within 95 % range 
- t-statistics 

**Type II error **
- predictive negative but actually True (FN).
- incorrectly retaining a false null hypothesis

## METRICS CALCULATIONS

|Test\Actual| True | False|
|-------|----|----|
|True|True Positve (TP)|false positive (FP)|
|False|False Negative (FN)|true negative (TN)|



$precision(P) = \frac{TP}{FP + TP}$

$recall(P) = \frac{TP}{TP+FN}$

$F-score = \frac{2*P*R}{P+R}$ 

$Accuracy = \frac{TP+TN}{TP+FP+TN+FN} $

source: https://medium.com/greyatom/performance-metrics-for-classification-problems-in-machine-learning-part-i-b085d432082b)
