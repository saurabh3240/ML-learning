```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv('pokemon-challenge/pokemon.csv')
print (data.info())
```

```python
#correlation map
f,ax = plt.subplots(figsize=(20,20))
sns.heatmap(data.corr(),annot=True,linewidths=.1,fmt='.1f',ax=ax)
```

```python
#line plot
plt.figure(figsize=(16,9))


data.Speed.plot(kind='line',color = 'g',label='Speed',linewidth =1,alpha=0.5,grid=True,linestyle=':')
data.Defense.plot(color = 'r',label='Defense',linewidth =1,alpha=0.5,grid=True,linestyle='-.')
plt.legend(loc='upper right')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Line Plot')
```

```python
#scatter plot
# x = attack y = defense
plt.figure(figsize=(20,20))
data.plot(kind='scatter',x='Attack',y ='Defense',alpha =0.5 ,color ='blue')
plt.xlabel('Attack')
plt.ylabel('Defense')
plt.title('Attack Defense Scatter Plot')
```

```python
#histogram
data.Speed.plot(kind='hist',bins=50,figsize=(15,15))
plt.xlabel('speed')
```
